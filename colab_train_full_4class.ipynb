```python
from pathlib import Path
import yaml

yaml_path = Path('/content/dataset/agriculture_vision_full_4class/dataset.yaml')
config = yaml.safe_load(yaml_path.read_text(encoding='utf-8'))
config['path'] = '/content/dataset/agriculture_vision_full_4class'
yaml_path.write_text(yaml.safe_dump(config, sort_keys=False), encoding='utf-8')
print(yaml_path.read_text())
```

```python
from ultralytics import YOLO

model = YOLO('yolov10s.pt')
results = model.train(
    data='/content/dataset/agriculture_vision_full_4class/dataset.yaml',
    epochs=80,
    imgsz=512,
    batch=16,
    device=0,
    patience=20,
    project='/content/runs',
    name='agriculture_yolov10s_4class',
)
```
